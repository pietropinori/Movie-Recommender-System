##  Rapid Automatic Keyword Extraction algorithm and Uses stopwords for english from NLTK, and all puntuation characters by default

```python
#imprting libraries

import collections                                    
import operator                                              
from rake_nltk import Rake
```

```python
#extracting top N keywords from movie overview using RAKE algorithm and assigning them a score.

def keywords_extracted (movie_overview, top_n_keywords=5):
    r = Rake()
    r.extract_keywords_from_text(movie_overview.lower())
    key_words_dict_scores = r.get_word_degrees()
    sorted_key_words_dict_scores = sorted(key_words_dict_scores.items(), key=operator.itemgetter(1), reverse=True)
    sorted_dict = collections.OrderedDict(sorted_key_words_dict_scores)
    
   
    return sorted_dict, list(sorted_dict.keys())[:top_n_keywords]
```

```python
#appling the above function

df2_movies['text_mined_overview']=df2_movies['overview'].apply(lambda x: keywords_extracted(x)[1]).apply(' '.join)
```

## Using the TfidfVectorizer, and cosine similarity technique

```python
#importing libraries

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#initializing a TF-IDF vectorizer with English stop words

tfidf = TfidfVectorizer(analyzer='word', stop_words='english')
```

```python
#creating a sparse matrix of TF-IDF scores for movie overviews

overview_matrix=tfidf.fit_transform(df_joinmovies['overview_all'])
```

```python
#calculating cosine similarities between movie overviews

cosine_similarities=cosine_similarity(overview_matrix)
```

```python
#finding the row in the df corresponding to 'Avatar'

TEST_MOVIE_NAME = 'Avatar'
TEST_MOVIE_INDEX = df_joinmovies.loc[df_joinmovies['title']==TEST_MOVIE_NAME].index.values[0].astype(int).item()

df_joinmovies.loc[df_joinmovies['title']==TEST_MOVIE_NAME]
```

```python
#returning indices of movies most similar to a given movie

cosine_similarities[TEST_MOVIE_INDEX].argsort()[:-7:-1]
```

```python
#calculating similar movies upon cosine similarities

similarities = {}
for i in range(len(cosine_similarities)):
    indices = cosine_similarities[i].argsort()[:-7:-1]
    similarities[df_joinmovies['title'].iloc[i]] = [( df_joinmovies['title'][x])
                                           for x in indices][1:]
    
```
